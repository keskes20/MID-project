```python
import numpy as np
import pandas as pd
import openpyxl
from IPython.display import display, HTML
```

```python
df1 = pd.read_excel('data\시도별_1인당_지역내총생산__지역총소득__개인소득_20250528134501.xlsx', engine='openpyxl')
df1.columns = [col.replace(' p)', '') for col in df1.columns]
display(HTML(df1.to_html(index=False)))
```

```python
df2 = df1[['시도별','2020','2021','2022']]
df2 = df2.iloc[1:]
df2
```

```python
# 수도권 지역 리스트 정의
capital_region = ['서울특별시', '인천광역시', '경기도']

# 새로운 열 '지역구분' 추가
df2['지역구분'] = df2['시도별'].apply(lambda x: '수도권' if x in capital_region else '비수도권')
df2
```

```python
df_grouped=df2.groupby('지역구분')[['2020', '2021', '2022']].mean()
df_grouped.to_excel('GRDP.xlsx', index=True)
```
